# Analysis of CRDC referral and arrest data

```python
import numpy as np
import pandas as pd
```

```python
enr = pd.read_csv("input/enrollment.csv", low_memory=False)
ref = pd.read_csv("input/referrals.csv", low_memory=False)
arr = pd.read_csv("input/arrests.csv", low_memory=False)
cha = pd.read_csv("input/characteristics.csv", low_memory=False)
```

```python
HVPLOT_ARGS = {
    "backend": "hvplot",
    "width": 900,
    "height": 600,
}
```

```python
def add_column_level(df, name):
    df.columns = pd.MultiIndex.from_tuples([(name, col) for col in df.columns])
    return df


def calculate_without_disabilities(df, name):
    df[(name, "without_disabilities")] = df[("enrollment", "overall")] - (
        df[("enrollment", "idea")] + df[("enrollment", "section_504")]
    )
    return df.drop((name, "overall"), axis=1)


def div_multiindex(df, name):
    return df.assign(**{name: lambda df: df[name] / df.enrollment * 1000})
```

## Disability status

```python
disability_status = (
    # crosstab for enrollment by disability status
    pd.crosstab(
        index=enr.combokey,
        columns=enr.variable,
        values=enr.enrollment,
        aggfunc=sum,
    )
    .pipe(add_column_level, "enrollment")
    # calculate total students without disabilities because
    # enrollment only has specific disabilities and "overall"
    .pipe(calculate_without_disabilities, "enrollment")
    # join referrals and arrests
    .join(
        pd.crosstab(
            index=arr.combokey,
            columns=arr.variable,
            values=arr.arrests,
            aggfunc=sum,
        ).pipe(add_column_level, "arrests")
    )
    .join(
        pd.crosstab(
            index=ref.combokey,
            columns=ref.variable,
            values=ref.referrals,
            aggfunc=sum,
        ).pipe(add_column_level, "referrals")
    )
    # drop any null values, which are an indication that one or more fields is missing data
    .dropna(how="any")
)

disability_status
```

### National rates

```python
disability_status_national = (
    disability_status.sum()
    .to_frame()
    .transpose()
    .pipe(div_multiindex, "arrests")
    .pipe(div_multiindex, "referrals")
    .drop("enrollment", axis=1)
)

disability_status_national.index = ["United States"]

disability_status_national
```

```python
(
    disability_status_national.stack()
    .droplevel(0)
    .transpose()
    .plot(
        kind="bar",
        title="Referrals and arrests per 1,000 students",
        # **HVPLOT_ARGS,
    )
)
```

### By state

```python
idea_vs_nondis = (
    lambda df, col: (
        (df[(col, "idea")] - df[(col, "without_disabilities")])
        / df[(col, "without_disabilities")]
    )
    .multiply(100)
    .to_frame((col, "idea_vs_nondis"))
)

disability_status_state = (
    disability_status.join(
        cha.set_index("combokey")[["lea_state_name"]].pipe(
            add_column_level, "characteristics"
        )
    )
    .groupby(("characteristics", "lea_state_name"))
    .sum()
    .pipe(div_multiindex, "arrests")
    .pipe(div_multiindex, "referrals")
    .drop("enrollment", axis=1)
    .append(disability_status_national)
    .pipe(lambda df: df.join(
        idea_vs_nondis(df, "arrests")
    ))
    .pipe(lambda df: df.join(
        idea_vs_nondis(df, "referrals")
    ))
)

disability_status_state.round(2)
```

## Race


```python
race = (
    # crosstab for enrollment by disability status
    pd.crosstab(
        index=enr.combokey,
        columns=enr.race,
        values=enr.enrollment,
        aggfunc=sum,
    )
    .pipe(add_column_level, "enrollment")
    .join(
        pd.crosstab(
            index=arr.combokey,
            columns=arr.race,
            values=arr.arrests,
            aggfunc=sum,
        ).pipe(add_column_level, "arrests")
    )
    .join(
        pd.crosstab(
            index=ref.combokey,
            columns=ref.race,
            values=ref.referrals,
            aggfunc=sum,
        ).pipe(add_column_level, "referrals")
    )
    # drop any null values, which are an indication that one or more fields is missing data
    .dropna(how="any")
)

race
```

### National

```python
race_national = (
    race.sum()
    .to_frame()
    .transpose()
    .pipe(div_multiindex, "arrests")
    .pipe(div_multiindex, "referrals")
    .drop("enrollment", axis=1)
)

race_national.index = ["United States"]

race_national
```

### By state

```python
black_vs_white = (
    lambda df, col: (
        (df[(col, "black")] - df[(col, "white")])
        / df[(col, "white")]
    )
    .multiply(100)
    .to_frame((col, "black_vs_white"))
)

state_rates = (
    race.join(
        cha.set_index("combokey")[["lea_state_name"]].pipe(
            add_column_level, "characteristics"
        )
    )
    .groupby(("characteristics", "lea_state_name"))
    .sum()
    .pipe(div_multiindex, "arrests")
    .pipe(div_multiindex, "referrals")
    .drop("enrollment", axis=1)
    .append(race_national)
    .pipe(lambda df: df.join(
        black_vs_white(df, "arrests")
    ))
    .pipe(lambda df: df.join(
        black_vs_white(df, "referrals")
    ))
)

state_rates.round(2)
```
